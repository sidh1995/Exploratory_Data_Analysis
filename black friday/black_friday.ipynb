```python
# import  the test data

df_test = pd.read_csv('C:/Users/Atul Vyas/Desktop/data analysis with python/cris_project/black friday/test.csv') 
df_test.head()
```

```python
# merge train and test data set

merge_dataset =  df_test.append(df_train)
merge_dataset.head()
```

```python
merge_dataset.info()
```

```python
merge_dataset.describe()
```

```python
merge_dataset.drop(['User_ID'],axis=1,inplace=True)
```

```python
merge_dataset.head()
```

```python
# another way to change gender column categorical to numerical dataset

# merge_dataset['Gender']= pd.get_dummies(merge_dataset['Gender'],drop_first= 1)
```

```python
# Heandling Categorical Feture  Gender

merge_dataset['Gender'] = merge_dataset['Gender'].map({'F' : 0, 'M' : 1})
merge_dataset
```

```python
# Handle Categorical Feture Age

merge_dataset['Age'].unique()
```

```python
# pd.get_dummies(merge_dataset['Age'],drop_first= True)

merge_dataset['Age'] = merge_dataset['Age'].map({'0-17' : 1, '18-25' : 2,'26-35' : 3,'36-45' : 4, '46-50' : 5,
                                                  '51-55' : 6, '55+' : 7})


# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
merge_dataset['Age']= label_encoder.fit_transform(merge_dataset['Age'])

merge_dataset['Age'].unique()
```

```python
merge_dataset.head() 
```

```python
# fixing categorical city_category

city_01= pd.get_dummies(merge_dataset['City_Category'],drop_first= True)
```

```python
city_01.head()
```

```python
merge_dataset = pd.concat([merge_dataset,city_01],axis =1)
```

```python
merge_dataset
```

```python
# drop city category

merge_dataset.drop('City_Category',axis=1,inplace= True)
```
